# Folium Example

```python
import folium as fl
```

## Get a Map from its Geocode

```python
myGeocode = (37.5101935, 127.0585824)
mapObject = fl.Map(location=myGeocode)  # COEX
mapObject
```

### You can Zoom it ( 1 ~ 19, default=10)

```python
mapObject = fl.Map(location=myGeocode, zoom_start=17)
mapObject
```

```python
mapObject = fl.Map(location=myGeocode, zoom_start=17, tiles='Stamen Toner')
mapObject = fl.Map(location=myGeocode, zoom_start=12, tiles='Stamen Terrain')
mapObject
```
